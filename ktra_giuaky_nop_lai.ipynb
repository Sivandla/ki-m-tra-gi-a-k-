```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. Tạo Dataframe như dữ liệu đã cho trên.

```python

data = {
    'Name': ["Mr. Harris", "Mrs. Bradley", "Miss. Laina",
    "Mrs. Heath", "Mr. William", "Mr. James", "Mr. Timothy",
    "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite", "Mr. William",
    "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],

    'Height': [1.8, 1.61, 1.78, 1.74, 1.55,
    1.9, 1.61, 1.4, 1.92, 1.77, 1.55, 1.85,
    1.64, 1.77],

    'Weight': [90, 87, 64, 70, 100, 106,
    82, 55, 77, 60, 100, 109, 80, 66],

    'AGE': [32, 45, 22, 52, 48, 21, 28,
    69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)
print("DataFrame ban đầu:")
print(df)
```

12. Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng
trùng lặp, giữ lại hàng đầu tiên

```python
df.drop_duplicates(subset='Name', keep='first', inplace=True)
print("DataFrame sau khi loại bỏ các hàng trùng lặp:")
print(df)
```

13. Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó

```python
mean_age = df['AGE'].mean()
df.loc[df['AGE'] > 40, 'AGE'] = mean_age
print("DataFrame sau khi thay thế các giá trị lớn hơn 40 bằng giá trị trung bình:")
print(df)
```

14. Biết công thức tính chỉ số BMI như sau:
BMI = Weight(kg)/Height2 (m2)  
Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BIM mỗi người

```python
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
print("DataFrame sau khi thêm cột cs_BMI:")
print(df)
```

15. Theo thang đo kết quả tính BMI của châu Á:
Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình  
cân, từ 25 trở lên là béo phì.
thường, 23-24,9 là thừa
Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ
số BMI trên.

```python
df['Phan_Loai'] = pd.cut(df['cs_BMI'], bins=[-float('inf'), 18.5, 23, 25, float('inf')],
                         labels=['0', '1', '2', '3'])
print("DataFrame sau khi thêm cột Phan_Loai:")
print(df)
```

16. Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1

```python
df['cs_BMI'] = (df['cs_BMI'] - df['cs_BMI'].min()) / (df['cs_BMI'].max() - df['cs_BMI'].min())
print("DataFrame sau khi chuẩn hóa cột cs_BMI:")
print(df)
```

17. Cho biết:
+ ‘Mr‘tương ứng với ‘male’,
+ ’Mrs’ và ‘Miss’ tương ứng với ‘female’.  
Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong
cột <Name> còn các từ đó.

```python
df['GT'] = df['Name'].apply(lambda x: 'male' if 'Mr' in x else ('female' if 'Mrs' in x or 'Miss' in x else x))
df['GT'] = df['GT'].apply(lambda x: x.replace('Mr', '').replace('Mrs', '').replace('Miss', ''))

df['Name'] = df['Name'].str.replace('Mr.', '')
df['Name'] = df['Name'].str.replace('Mrs.', '')
df['Name'] = df['Name'].str.replace('Miss.', '')

print("DataFrame sau khi thêm cột GT:")
print(df)
```

18. Vẽ biểu đồ tròn cho <GT>.

```python
plt.figure(figsize=(5, 5))
df['GT'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Biểu đồ tròn cho cột GT')
plt.ylabel('')
plt.show()
```

19. Trực quan hóa dữ liệu người nữ trên <Phan_Loai>.

```python
df_female = df[df['GT'] == 'female']
df_female['Phan_Loai'].value_counts().plot(kind='bar')
plt.xlabel('Phan_Loai')
plt.ylabel('Count')
plt.title('Trực quan hóa dữ liệu người nữ trên Phan_Loai')
plt.show()
#
```
